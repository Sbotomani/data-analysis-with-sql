### 2. A Palindrome is a string that reads the same backwards and forwards. Write a Python program to determine whether or not a given string is a Palindrome. [10]

```python
def is_palindrome(s):

    s = s.replace(" ", "").lower()
   
    return s == s[::-1]


input_string = input("The entered string: ")


if is_palindrome(input_string):
    print(input_string,' is a palindrome.')
else:
    print(input_string,' is not a palindrome.')
```

### 8. Use Linear Regression to predict the loan category of a client based on their income. Use 75% as training data and 25% as testing data. Show all steps clearly and include comments. [15]

```python
df['Loan Category']=pd.to_numeric(df['Loan Category'], errors = 'coerce')
average_age = df['Loan Category'].mean()
df['Loan Category'].fillna(average_age, inplace = True)
```

```python
# import necessary packages
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
# The empty regressor into the model 'ORTS'
ORTS = LinearRegression()
ORTS
```

```python

# fit the regressor into the model 'ORTS'
ORTS.fit(Income_train, Loan_train)
```

```python
# pridict the variables given the real variables
predict_Loan_test = ORTS.predict(Income_test)
predict_Loan_train = ORTS.predict(Income_train)
```
